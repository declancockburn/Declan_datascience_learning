```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
%matplotlib inline 
```

```python
path='https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/automobileEDA.csv'
df = pd.read_csv(path)
```

```python
sns.regplot(x=df['highway-mpg'],y=df['price'])
```

```python
sns.residplot(x=df['highway-mpg'],y=df['price'])
```

```python
lm = LinearRegression()
X = df[['highway-mpg']]
Y = df['price']
lm.fit(X, Y)
Yhat = lm.predict(X)
```

```python
ax1 = sns.distplot(df['price'], hist=False, color='r',
                  label='Actual Value')

sns.distplot(Yhat, hist=False, color="b", 
             label='Fitted Values', ax=ax1)
```

```python
df2 = df
df2['Yhat'] = Yhat
df2.Yhat
```

```python
df2 = df
df2['Yhat'] = Yhat
ax2 = sns.displot(x='price', kind='kde', color='r',
                  label='Actual Value', data=df2)
sns.displot(x='Yhat', kind='kde', color="b", 
             label='Fitted Values', data=df2, ax=ax2)
```

```python
ax2 = sns.kdeplot(x='price', color='r',
                  label='Actual Value', data=df2)
sns.kdeplot(x='Yhat', color="b", 
             label='Fitted Values', data=df2, ax=ax2)
```
