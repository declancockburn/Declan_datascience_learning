```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
%matplotlib inline 
```

```python
path='https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/automobileEDA.csv'
df = pd.read_csv(path)
```

```python
sns.regplot(x=df['highway-mpg'],y=df['price'])
```

```python
sns.residplot(x=df['highway-mpg'],y=df['price'])
```

```python
lm = LinearRegression()
X = df[['highway-mpg']]
Y = df['price']
lm.fit(X, Y)
Yhat = lm.predict(X)
```

```python
ax1 = sns.distplot(df['price'], hist=False, color='r',
                  label='Actual Value')

sns.distplot(Yhat, hist=False, color="b", 
             label='Fitted Values', ax=ax1)
```

```python
df2 = df
df2['Yhat'] = Yhat
df2.Yhat
```

```python
df2 = df
df2['Yhat'] = Yhat
ax2 = sns.displot(x='price', kind='kde', color='r',
                  label='Actual Value', data=df2)
sns.displot(x='Yhat', kind='kde', color="b", 
             label='Fitted Values', data=df2, ax=ax2)
```

```python
ax2 = sns.kdeplot(x='price', color='r',
                  label='Actual Value', data=df2)
sns.kdeplot(x='Yhat', color="b", 
             label='Fitted Values', data=df2, ax=ax2)
```

```python
# Multiple linear regression:
lm = LinearRegression() # Blank model
# Data
Z = df[['horsepower', 'curb-weight', 'engine-size', 'highway-mpg']]
# Create model
lm.fit(Z, df['price']) 

# Now check how model predicts itself, distribution plot:
Y_hat = lm.predict(Z)
```

```python
# Plot kernal density plot (distribution plot!) to see how accurate the fit is.

plt.figure(figsize=(width, height))

#ax1 = sns.distplot(df['price'], hist=False, color="r", label="Actual Value")
#sns.distplot(Y_hat, hist=False, color="b", label="Fitted Values" , ax=ax1)

df['Yhat']=Y_hat
ax1 = sns.kdeplot(x='price', color='r', label='Actual Value', data=df)
sns.kdeplot(x='Yhat', color="b", label='Fitted Values', data=df, ax=ax1)

plt.title('Actual vs Fitted Values for Price')
plt.xlabel('Price (in dollars)')
plt.ylabel('Proportion of Cars')

plt.show()
plt.close()

```
